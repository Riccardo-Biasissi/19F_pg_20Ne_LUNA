```python
# Read the CSV file of yield for the target
df = pd.read_csv(f"/data0/biasissi/LUNA/19F+p_g+20Ne/Scripts/Data/Yield/Yield_csv/Yield_{target}.csv")

# print(df)

# Count the number of scans in the dataframe with E that are not NaN and E<300 keV
num_scans = 0
for col in df.columns:
    if col.startswith("E_scan"):
        if df[col].notna().any() and (df[col] < 300).any():
            num_scans += 1
print(f"Number of scans in the data: {num_scans}")

scans = np.arange(1, num_scans + 1)
scans = [4]

for scan in scans:
    # Recover the very first scan
    x, y, y_err = df[f"E_scan{scan}"], df[f"Y_scan{scan}"], df[f"Y_error_scan{scan}"]

    # Refine the yield data by selecting only points with E not NaN and E<300 keV
    mask = (~np.isnan(x)) & (x < 300)
    x = x[mask]
    y = y[mask]
    y_err = y_err[mask]

x.to_numpy()
y.to_numpy()
y_err.to_numpy()

eff = 0.689
q_e = 1.602176634e-19

# N_reactions / N_protons
#y = y * q_e / eff
#yerr = yerr * q_e / eff

# Convert to CM
x = x * 19 / 20.007

# Read cross-section
data = np.loadtxt( "utils/19F_pag_200-260keV.extrap", usecols=(0,3))
energies, extrap = data[:,0], data[:,1]
```
